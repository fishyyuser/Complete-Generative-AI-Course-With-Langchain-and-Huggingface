```python
import os
from dotenv import load_dotenv
load_dotenv()
os.environ["OPENAI_API_KEY"]=os.getenv("OPENAI_API_KEY")
os.environ["LANGCHAIN_API_KEY"]=os.getenv("LANGCHAIN_API_KEY")
os.environ["LANGCHAIN_TRACING_V2"]="true"
os.environ["LANGCHAIN_PROJECT"]=os.getenv("LANGCHAIN_PROJECT")
```

```python
from langchain_openai import ChatOpenAI
llm=ChatOpenAI(model="gpt-5-nano")
print(llm)
```

```python
### Input and get response from LLM
query="What is Generative AI?"
response=llm.invoke(query)
```

```python
print(response.content)
```

```python
response=llm.invoke("What is langchain and how is it used by developers?")
print(response.content)
```

```python
### chat prompt template
from langchain_core.prompts import ChatPromptTemplate
```

```python
prompt=ChatPromptTemplate.from_messages(
    [
        ("system","You are an expert AI/ML engineer. You need to answer few questions"),
        ("user","{input}")
    ]
)
```

```python
chain=prompt|llm
response=chain.invoke({"input":"what jobs are going to go extinct in next 5 years let's say 2030 and what new jobs will be created?"})
```

```python
text=response.content.split(".")
for line in text:
    print(line)
```
