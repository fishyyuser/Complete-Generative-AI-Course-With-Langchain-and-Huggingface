```python
# building the db
from langchain_chroma.vectorstores import Chroma
```

```python
from langchain_community.document_loaders import TextLoader
from langchain_ollama import OllamaEmbeddings
from langchain_text_splitters import CharacterTextSplitter
```

```python
docs=TextLoader("speech2.txt").load()
docs
```

```python
textSplitter=CharacterTextSplitter(chunk_size=1000,chunk_overlap=30)
dox=textSplitter.split_documents(docs)
print(dox)
```

```python
embedding=OllamaEmbeddings(model="embeddinggemma")
db=Chroma.from_documents(embedding=embedding,documents=dox)
```

```python
query="How meaningful progress happens"
query_response=db.similarity_search(query)
```

```python
for doc in query_response:
    print(doc.page_content,end="\n")
```

```python
retreiver=db.as_retriever()
query_response=retreiver.invoke(query)
```

```python
for doc in query_response:
    print(doc.page_content,end="\n")
```

```python
docs_and_score=db.similarity_search_with_score(query=query)
```

```python
print(docs_and_score[0])
```

```python
embedded_vector=embedding.embed_query(query)
embedded_vector
```

```python
db.similarity_search_by_vector(embedded_vector)
```

```python
# save in local
vectordb=Chroma.from_documents(documents=dox,embedding=embedding,persist_directory="./chroma_db")
```
